# 1.What percentage of all licensed sets ever released where star war themed?
### first merge two sheets
### find total that have license
### then find how many star war licensed
### then find the percentage of star wars in licensed


```python
merged=data.merge(theme, left_on='parent_theme', right_on='name')

merged.head(50)
```

```python
licensed=merged[merged['is_licensed']]
licensed.head(50)
count_license=licensed.shape[0]
print(count_license)
```

```python
starwars=licensed[licensed['parent_theme']=='Star Wars']
starwars.head(50)
count_star=starwars.shape[0]
print(count_star)
```

```python
percent=int((count_star/count_license)*100)
print (percent)
```

# 2. In which year was Star wars not the most famous licensed theme(In terms of number of themes released that year)? 
### first group the licensed data with year and parent_theme
### sum the data to use 'is_licensed' to use as total count
### sort and drop duplicates to find maximum in each year

```python
licensed=licensed.sort_values(by=['year'])
```

```python

sorted_data=licensed.groupby(['year','parent_theme'])
sorted_data=sorted_data.sum(numeric_only=True).reset_index()
# sorted 'is_licensed to find max'
find_max=sorted_data.sort_values('is_licensed', ascending=False)
#drop duplicates to find max in every year
find_max.drop_duplicates('year')
```

```python
Year_StarWars_Not_Maximum=2017
```

### 2. In which year was Star wars not the most famous licensed theme(In terms of number of themes released that year)?
### 2017
